# **Data Distribution**


#### **Main types of data distributions:**
1. **Normal Distribution**: Continuous, bell-shaped curve. Characterized by mean and standard deviation. Used in many sciences.

2. **Uniform Distribution**: All outcomes equally likely. Example: deck of cards.

3. **Binomial Distribution**: Discrete, number of successes in n trials. Characterized by number of trials and probability of success. Used when two mutually exclusive outcomes.

4. **Poisson Distribution**: Discrete, number of events in fixed interval. Characterized by average number of events. Used for modelling number of times an event occurs.

5. **Exponential Distribution**: Continuous, time between events in Poisson process. Characterized by rate parameter. Used to model lifespan or time between events.

# **Distributions**


```python
import numpy as np
import matplotlib.pyplot as plt

# Generating example datasets with different distributions
np.random.seed(0)  # For reproducibility

# Normal distribution data
normal_data = np.random.normal(loc=0, scale=1, size=1000)

# Uniform distribution data
uniform_data = np.random.uniform(low=-2, high=2, size=1000)

# Exponential distribution data
exponential_data = np.random.exponential(scale=1, size=1000)

# Plotting the distributions
plt.figure(figsize=(12, 8))

# Normal Distribution Plot
plt.subplot(3, 1, 1)
plt.hist(normal_data, bins=30, color='skyblue', alpha=0.7)
plt.title('Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Uniform Distribution Plot
plt.subplot(3, 1, 2)
plt.hist(uniform_data, bins=30, color='lightgreen', alpha=0.7)
plt.title('Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Exponential Distribution Plot
plt.subplot(3, 1, 3)
plt.hist(exponential_data, bins=30, color='salmon', alpha=0.7)
plt.title('Exponential Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

#### **Important Questions:**
**`1. Why normal distrobution is importrant in Data Science/Data Analysis? (Parametric Tests / Non-Parametric Tests)`**

Normal distribution is important in Data Science and Data Analysis for several reasons:

- **Central Limit Theorem**: The sum of many independent random variables, regardless of their distribution, tends to follow a normal distribution. This makes normal distribution a common model for many natural phenomena.

- **Parametric Tests**: Many statistical tests assume normal distribution (e.g., t-test, ANOVA). These tests are more powerful (i.e., more likely to reject the null hypothesis when it is false) than their non-parametric counterparts when the assumption of normality is met.

- **Ease of Computation**: Normal distribution simplifies the computation and interpretation of probabilities.

- **Symmetry**: Normal distribution is symmetric, which simplifies the analysis and interpretation of data.

- **Predictive Modeling**: Many machine learning algorithms assume that the input variables are normally distributed, which can lead to better performance of the model.

**`2. What are some common non-parametric tests used in Data Science and Data Analysis?`**

Non-parametric tests are statistical methods that do not assume a specific distribution to the data. Here are some commonly used non-parametric tests in Data Science and Data Analysis:

- **Mann-Whitney U Test**: Used to compare differences between two independent groups when the dependent variable is either ordinal or continuous, but not normally distributed.

- **Wilcoxon Signed-Rank Test**: Used to compare paired samples, similar to the paired t-test, but without the assumption of normal distribution.

- **Kruskal-Wallis H Test**: Used to compare more than two independent groups of ordinal or continuous data. It's the non-parametric version of ANOVA.

- **Spearman's Rank Correlation**: Used to measure the strength and direction of association between two ranked variables.

- **Chi-Square Test of Independence**: Used to determine if there is a significant association between two categorical variables.

- **Kolmogorov-Smirnov Test**: Used to compare a sample with a reference probability distribution, or to compare two samples.

- **Fisher's Exact Test**: Used to analyze contingency tables when sample sizes are small.
